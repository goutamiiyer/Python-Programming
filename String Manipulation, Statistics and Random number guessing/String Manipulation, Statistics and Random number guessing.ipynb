# Question 1 - Rock Paper Scissors Game.

## High Level Algorithm:

1. Receive the user's choice of Rock, Paper, Scissors
2. The program or computer randomly chooses one of the three choices from the list created.
3. Program or computer compares its random choice with the user's choice.
    3.1. Rock smashes scissors.
    3.2. Scissors cuts paper.
    3.3. Paper covers rock
4. The output shows the userâ€™s choice, the programâ€™s choice and the winner.

## Instructions:

1. Run the cell containing the program (Alt+Enter).
2. Enter the the choice and press Enter.
3. Results are displayed.

```python
# Question 1

from random import randint
# Driver code
def main():
    option = True
    
    while(option):
        # Receive the user's choice as input and validate.
        user = input("Enter your choice: ")
        if user == "rock" or user == "paper" or user == "scissors":
            pass
        else:
            print("Enter a valid input. Your input may have spelling errors.")
            option = False
            break;

        # Create a list with all the choices.
        choiceList = ["rock", "paper", "scissors"]

        # Assign a random choice to the computer.
        computer = choiceList[randint(0,2)]

        # Display the computer's choice.
        print("Computer chooses " + computer)

        # We apply the rules of this game after we get user's choice.
        if user == computer:
            print("User and computer tie!")
        elif user == "rock":
            # Paper covers rock.
            if computer == "paper":
                print("I, the computer, win!")
            else:
                print("You, the user, win!")
        elif user == "paper":
            # Scissors cuts paper.
            if computer == "scissors":
                print("I, the computer, win!")
            else:
                print("You, the user, win!")
        elif user == "scissors":
            # Rock smashes scissors.
            if computer == "rock":
               print("I, the computer, win!")
            else:
                print("You, the user, win!")
        else:
            print("This is not a valid input.")

        option = False

# We define the name of the main function.
if __name__ == "__main__":
    main()
```

# Question 2 - Number guessing game.

## High Level Algorithm:

1. Receive the number to be guessed from the user between 1 and 100(both inclusive) and validate the input
2. The program or computer is asked to guess the number.
3. The program is looped until it guesses the number given by the user and for each iteration a count is maintained. 
4. When the program or computer guesses the number, output is displayed.

## Instructions:

1. Run the cell containing the program (Alt+Enter).
2. Enter number to be guessed and press Enter.
3. Results are displayed.

```python
# Question 2

import random

# Driver code
def main():
    while(True):
        try:
            # Receive the number to be guessed from the user and validate.
            user = int(input("Enter number to be guessed: "))
        except ValueError:
            print("Expected input was only integer")
        # Check if the user's input number is between 1 and 100, both inclusive.
        if user >= 1 and user <= 100:
            break
        else:
            print("Sorry, the input should be between 1 and 100, both inclusive")

    # Assign a random choice to the computer.
    computer = random.randint(1,100)
    # Set a counter to calculate the number of iterations it takes for the computer to guess the number.
    count = 1
    
    # Run in a loop until the computer guesses the user's number.
    while(computer != user):
        # The computer guesses a random number between 1 and 100, inclusive.
        computer = random.randint(1,100)
        count += 1
    
    # After the computer has guessed the user's number, display the output.
    print("You guessed " + str(user) + ", and it took the program " + str(count) + " iterations")
    

# We define the name of the main function.
if __name__ == "__main__":
    main()
```

# Question 3 - The number of times the letter is the first character of a word.

## High Level Algorithm:

1. Receive the input line of text from the user.
2. Receive the letter or character to be searched in the line of text.
3. For every word in the line, it is checked if the first character is equal to the letter enetered by the user.
    3.1. If it is equal, then we count the number of times it occurs in the line.
    3.2. If not we just display and come out of the loop

## Instructions:

1. Run the cell containing the program (Alt+Enter).
2. Enter the line of text and press Enter.
3. Enter the letter to be searched and press Enter.
4. Results are displayed.

```python
# Question 3

# Driver code
def main():
    # Receive the input line of text from the user.
    line = input("Enter your line of text: ")
    # Receive the letter to be searched from the line of text.
    letter = input("Enter your letter to use: ")
    
    # We create a list which contains all the words in the line of text enetered.
    listOfWords = line.split()
    # Set a counter to count the number of times the letter occurs as the first letter in a word.
    count = 0
    
    # Iterate through the list of words
    for i in range(len(listOfWords)):
        # Check if the first letter in the word starts with the letter enetered by the user.
        if (listOfWords[i].startswith(letter)):
            count += 1
        else:
            continue

    # Display the number of times the letter occurs as the first letter in each word in the text.
    print("Your letter "+ letter +" occurs as the first letter: " + str(count) +" times")

# We define the name of the main function.
if __name__ == "__main__":
    main()
```

# Question 4 - Determine whether a string is a palindrome or not.

## High Level Algorithm:

1. Receive the input string which can be a word or a sentence.
2. Check if it is a word.
    2.1. If it is a word, then call the function wordPalindrome()
    2.2. We loop through the string and check if first letter equals the last letter, second letter equals the second last letter and so on.
    2.3. If everything is equal, display thet it is a palindrome else it is not. 
    2.4. Return to the main function.
3. Check if it is a sentence.
    3.1. If it is a sentence, we set two pointers, one to the left and other to the right.
    3.2. We make all the letters in the sentence to lower case in order to make it easy for comparison.
    3.3. Loop until the two pinters from left and right meet and check if the characters are equal ignoring spaces
    3.4. If the characters are equal, sentence is a palindrome. Otherwise it is not a palindrome.

## Instructions:

1. Run the cell containing the program (Alt+Enter).
2. Enter the snetence or a word and press Enter.
3. Results are displayed.

```python
# Question 4 

# Function to find out if a word is a palindrome or not.
def wordPalindrome(inputString):
    # For every character in the first half of the word
    for i in range(0, len(inputString)//2):
        # Check if the first half is equal to the second half and return to main function
        if inputString[i] != inputString[len(inputString)-i-1]:
            print('"' + inputString + '" is not a palindrome')
            return
    print('"' + inputString + '" is a palindrome')
    return

# Function to find out if a sentence is a palindrome or not.
def sentencePalindrome(inputString):
    # Set two pointers, one to the left and other to the right.
    left, right = 0, len(inputString) - 1
    # All letters changed to lower case to make it easy for comparison.
    lowerString = inputString.lower() 
    
    # Loop until the two pinters from left and right meet.
    while(left <= right):
        # If there is another symbol in left of sentence.
        if (not(lowerString[left] >= 'a' and lowerString[left] <= 'z')):
            left += 1;
        # If there is another symbol in right of sentence.
        elif (not(lowerString[right] >= 'a' and lowerString[right] <= 'z')):
            right -= 1;
        # If characters are equal.
        elif (lowerString[left] == lowerString[right]):
            left += 1
            right -= 1
        else:
            # If characters are not equal then sentence is not palindrome.
            print('"' + inputString + '" is not a palindrome')
            return
    # prints and returns if sentence is a palindrome
    print('"' + inputString + '" is a palindrome')
    return

# Driver code
def main():
    # Receive the input string which is a word or a sentence.
    inputString = input("Enter a string: ")
    # We split the string for spaces.
    stringList = inputString.split()
    # If the string has more than 1 word, we consider it to be a sentence
    if len(stringList) > 1:
        sentencePalindrome(inputString)
    else:
        # Otherwise, it has only one word
        wordPalindrome(inputString)
        
# We define the name of the main function.
if __name__ == "__main__":
    main()
```

# Question 5 - Statistics of the contents of the file.

## High Level Algorithm:

1. Receive input file name from the user and validate the file name by opening it.
2. Counters are set to calculate  the number of lines, words and characters in the file.
3. For every line in the file, 
    3.1. Count the number of lines and words in each line.
    3.2. For every word in the words list, count the number of characters in each word.
4. Average Length of words is total number of characters divided by total number of words.
5. Display all the results obtained.

## Instructions:

1. Run the cell containing the program (Alt+Enter).
2. Enter the file name to be assessed and press Enter.
3. Results are displayed.

```python
# Question 5

# Driver code
def main():
    # Loop until the user enters the correct file name.
    while(True):
        # Receive the file name as input and validate.
        fileName = input("What is the filename: ")
        try:
            file = open(fileName,'r')
            break
        except:
            print("File cannot be opened: ", fileName)
            print("Enter the correct file name present in your directory")
    
    # Set the counter for number of lines, words and characters.
    lineCount = 0
    wordCount = 0
    charCount = 0
    # For every line in the file.
    for line in file:
        # Calculate the number of lines.
        lineCount += 1
        # Calculate the number of words.
        words = line.split()
        wordCount += len(words)
        # For every word in a sentence.
        for word in words:
            # Calculate the number of characters in each word.
            charCount += len(word)
    # Average Length of words is total number of characters divided by total number of words
    avgLength = charCount / wordCount
    # rounded to 1 decimal place
    avgLength = round(avgLength, 1)
    
    # Displaying the results
    print("Number of lines: " + str(lineCount))
    print("Number of words: " + str(wordCount))
    print("Number of characters: " + str(charCount))
    print("Average length of a word: " + str(avgLength))
    # Closing the file that was opened earlier
    file.close()

# We define the name of the main function.
if __name__ == "__main__":
    main()
```
