# Question 2 - Program to find average, maximum, minimum and median of 3 or 4 integers.

## High Level Algorithm:

1. Receive the number of integers (3 or 4) for which we need to calculate the average, maximum, minimum and median.
2. A separate function is written to calculate the maximum and minimum of three numbers. 
    2.1. Each number is checked if it is greater than other two numbers.
    2.2. Each number is checked if it is less than other two numbers.
3. A separate function is written to calculate the maximum and minimum of four numbers.
    3.1. Each number is checked if it is greater than other three numbers.
    3.2. Each number is checked if it is less than other three numbers.
4. Average is calculated by adding all the numbers and dividing by the number of integers(3 or 4).
5. Median is calculated by sorting all the input numbers and finding the middle term. If there are two numbers in the middle, we find the average of those two numbers.
    5.1. For the median, we create a list and add all the numbers to the list from the input.
    5.2. Sort the list of inputs and find the middle term
    5.3. If there are two middle terms, find average of those two numbers.

## Instructions:

1. Run the cell containing the program (Alt+Enter).
2. Enter the number of integers(3 or 4) and all the numbers one by one and press Enter.
3. Results are displayed.

```python
# Question 2
# Function to find the maximum and minimum of three numbers
def maxMinOfThree(firstNum, secondNum, thirdNum):
    # Maximum of three numbres
    # If the first number is greater than other 2 numbers, first number is the maximum
    if (firstNum >= secondNum) and (firstNum >= thirdNum):
        maximum = firstNum
    # If the seccond number is greater than other 2 numbers, second number is the maximum
    elif (secondNum >= firstNum) and (secondNum >= thirdNum):
        maximum = secondNum
    # Otherwise, third number is the maximum
    else:
        maximum = thirdNum
    
    # Minimum of three numbres
    # If the first number is less than other 2 numbers, first number is the minimum
    if (firstNum <= secondNum) and (firstNum <= thirdNum):
        minimum = firstNum
    # If the seccond number is less than other 2 numbers, second number is the minimum
    elif (secondNum <= firstNum) and (secondNum <= thirdNum):
        minimum = secondNum
    # Otherwise, third number is the minimum
    else:
        minimum = thirdNum
    return (maximum,minimum)

# Function to find the maximum and minimum of four numbers
def maxMinOfFour(firstNum, secondNum, thirdNum, fourthNum):
    # Maximum of four numbres
    # If the first number is greater than other 3 numbers, first number is the maximum
    if (firstNum >= secondNum) and (firstNum >= thirdNum) and (firstNum >= fourthNum):
        maximum = firstNum
    # If the seccond number is greater than other 3 numbers, second number is the maximum
    elif (secondNum >= firstNum) and (secondNum >= thirdNum) and (secondNum >= fourthNum):
        maximum = secondNum
    # If the third number is greater than other 3 numbers, third number is the maximum
    elif (thirdNum >= firstNum) and (thirdNum >= secondNum) and (thirdNum >= fourthNum):
        maximum = thirdNum
    # Otherwise, fourth number is the maximum
    else:
        maximum = fourthNum
    # Minimum of four numbres
    # If the first number is less than other 3 numbers, first number is the minimum
    if (firstNum <= secondNum) and (firstNum <= thirdNum) and (firstNum <= fourthNum):
        minimum = firstNum
    # If the seccond number is less than other 3 numbers, second number is the minimum
    elif (secondNum <= firstNum) and (secondNum <= thirdNum) and (secondNum <= fourthNum):
        minimum = secondNum
    # If the third number is less than other 3 numbers, third number is the minimum
    elif (thirdNum <= firstNum) and (thirdNum <= secondNum) and (thirdNum <= fourthNum):
        minimum = thirdNum
    # Otherwise, fourth number is the minimum
    else:
        minimum = fourthNum
    return (maximum,minimum)

# Driver code
def main():
    numList = []
    # Receive the inputs and validate them for errors.
    while(True):
        # Receive the number of integers required. Either 3 or 4.
        try:
            noOfIntegers = int(input("Enter set of 3 or 4 integers:"))
        except ValueError:
            print("\nExpected input was only integer")
            continue
        # Check if the input of noOfIntegers is 3 or 4.
        if (noOfIntegers == 3) or (noOfIntegers == 4):
            break
        else:
            print("Sorry, expected input was 3 or 4")

    while(True):
        # Receive the first number and validate.
        try:
            firstNum = int(input("Enter first number:"))
            numList.append(firstNum)
            break
        except ValueError:
            print("\nExpected input was only integer")

    while(True):
        # Receive the second number and validate.
        try:
            secondNum = int(input("Enter second number:"))
            numList.append(secondNum)
            break
        except ValueError:
            print("\nExpected input was only integer")

    while(True):
        # Receive the third number and validate.
        try:
            thirdNum = int(input("Enter third number:"))
            numList.append(thirdNum)
            break
        except ValueError:
            print("\nExpected input was only integer")

    if (noOfIntegers == 4):
        while(True):
            # Receive the fourth number and validate.
            try:
                fourthNum = int(input("Enter fourth number:"))
                numList.append(fourthNum)
                break
            except ValueError:
                print("Expected input was only integer")

    # Median of all the numbers in the numList list.
    n = len(numList)
    numList.sort()
    if n % 2 == 0:
        median1 = numList[n//2]
        median2 = numList[n//2 - 1]
        median = (median1 + median2)/2
    else:
        median = numList[n//2]

    # Display the results
    if (noOfIntegers == 3):
        average = (firstNum+secondNum+thirdNum)/noOfIntegers
        print("The average of " + str(firstNum) + ", " + str(secondNum) + " and " + str(thirdNum) + " is: " + str(average))
        maximum,minimum = maxMinOfThree(firstNum, secondNum, thirdNum)
        print("The maximum of the three numbers is: " + str(maximum))
        print("The minimum of the three numbers is: " + str(minimum))
        print("The median of the three numbers is:" + str(median))
    if (noOfIntegers == 4):
        average = (firstNum+secondNum+thirdNum+fourthNum)/noOfIntegers
        print("The average of " + str(firstNum) + ", " + str(secondNum) + ", " + str(thirdNum) + " and " + str(fourthNum) + " is: " + str(average))
        maximum,minimum = maxMinOfFour(firstNum, secondNum, thirdNum, fourthNum)
        print("The maximum of the four numbers is: " + str(maximum))
        print("The minimum of the four numbers is: " + str(minimum))
        print("The median of the four numbers is: " + str(median))

# We define the name of the main function.
if __name__ == "__main__":
    main()
```

# Question 3 - Program to check if the number is prime or not.

## High Level Algorithm:

1. Receive an input number and pass its value to a function to check if it is prime.
2. A separate function is written to check if a number is prime or not.
    2.1. If a number is only divisible by 1 and itself it is prime
    2.2. If it is divisible by other numbers before it is composite
    2.3. If the number is 0 or 1, it is neither prime nor composite

## Instructions:

1. Run the cell containing the program (Alt+Enter).
2. Enter the number to be checked and press Enter.
3. Results are displayed.

```python
# Question 3
# Function to check if number is prime or composite
def checkIfPrime(num):
    # If the received number is negative, display error.
    if num < 0:
        print("Error: You did not enter a positive integer negative")

    # If the input is 0 or 1, display output.
    if num == 0 or num == 1:
        print("It is neither prime nor composite")

    # If input greater than 1, we divide input number with all other numbers in the range.
    # If it divides without a remainder, then it is not prime. Otherwise it is prime.
    if num > 1:
        if (num == 2):
            print("It is a prime")
        for i in range(2,num):
            if (num % i == 0):
                print("It is a composite")
                break
            else:
                print("It is a prime")

# Driver code
def main():
    # User has to run this script for every input
    while(True):
        # Receive the positive number.
        try:
            num  = int(input("Enter positive integer: "))
            # Call the function to check prime
            checkIfPrime(num)
            break
        except ValueError:
            print("Error: You did not enter a positive integer")
            break

# We define the name of the main function.
if __name__ == "__main__":
    main()
```

# Question 4 - Program to calculate the simple interest

## High Level Algorithm:

1. Receive inputs for principle amount(float), interest rate(float) and years(int) and pass its value to a function to calculate the simple interest.
2. Multiply the principle amount, interest rate and years and divide the result by 100 to calculate the Simple Interest.

## Instructions:

1. Run both the cells containing the program (Alt+Enter). Run the cell containing SimpleInterest() function first before running the cell containing the main() function.
2. Enter the inputs such as principle amount, interest rate and years and press Enter.
3. Results are displayed.

```python
# Question 4
# Please run this cell before you run the next cell in this ipynb file
# Simple interest is the product of Principle amount, interest rate and number of years, all divided by 100
# Function to find the Simple interest
# function definition in one cell
def SimpleInterest(principleAmt, interestRate, years):
    simpleInterest = (principleAmt*interestRate*years)/100
    return simpleInterest
```

```python
# Question 4 continued
# Please run the previous cell in this ipynb file before running this cell
# Driver code
# function calling main program below it or in a different cell
def main():
    while(True):
        # Receive the third number and validate.
        try:
            principleAmt = float(input("Enter Principle Amount: $"))
            break
        except ValueError:
            print("Expected input was only float or integer")

    while(True):
        # Receive the third number and validate.
        try:
            interestRate = float(input("Enter the interest rate: "))
            break
        except ValueError:
            print("Expected input was only float or integer")

    while(True):
        # Receive the third number and validate.
        try:
            years = int(input("Enter the year: "))
            break
        except ValueError:
            print("Expected input was only integer")

    # To calculate simple interest, we call the function that calculates by getting 3 inputs.
    simpleInterest = SimpleInterest(principleAmt, interestRate, years)
    print("Simple interest amount is: $" + str(simpleInterest))

# We define the name of the main function.
if __name__ == "__main__":
    main()
```

# Question 5 - Program to find whether the point (x,y) lies within the rectangle.

## High Level Algorithm:

1. Receive inputs for the coordinates of the rectangle (x1,y1), (x2,y2) and get the point to be checked (x,y) and pass all the values to a function to check if the point lies within the rectangle.
2. Check if the point x lies between x1 and x2 i.e. bottom-right and top-left coordinates.
3. Check if the point y lies between y1 and y2 i.e. bottom-right and top-left coordinates.

## Instructions:

1. Run the cell containing the program (Alt+Enter).
2. Enter the coordinates x1,y1,x2,y2 and the points to be checked x,y and press Enter.
3. Results are displayed.

```python
# Question 5
# Function to find if a coordiante (x,y) lies inside a rectangle or not
# A point lies inside the rectangle if and only if it’s 
# x-coordinate lies between the x-coordinate of the given bottom-right and top-left coordinates of the rectangle and
# y-coordinate lies between the y-coordinate of the given bottom-right and top-left coordinates of the rectangle
def checkPointLies(x1, y1, x2, y2, x, y):
    if (x > x1 and x < x2 and y > y1 and y < y2):
        return True
    else:
        return False

# Driver code
def main():
    response = 'Y'
    # Keep receiving inputs until user wishes to continue
    while response == 'Y':
        # Receive the coordinates x1,y1,x2,y2,x and y as input and validate.
        # Receive the coordinate x1 as input and validate for errors
        while(True):
            try:
                x1 = float(input("Enter x1:"))
                break
            except ValueError:
                print("Error! Bookmark not defined.")

        # Receive the coordinate y1 as input and validate for errors
        while(True):
            try:
                y1 = float(input("Enter y1:"))
                break
            except ValueError:
                print("Error! Bookmark not defined.")

        # Receive the coordinate x2 as input and validate for errors
        while(True):
            try:
                x2 = float(input("Enter x2:"))
                break
            except ValueError:
                print("Error! Bookmark not defined.")

        # Receive the coordinate y2 as input and validate for errors
        while(True):
            try:
                y2 = float(input("Enter y2:"))
                break
            except ValueError:
                print("Error! Bookmark not defined.")

        # Receive the coordinate x as input and validate for errors
        while(True):
            try:
                x = float(input("Enter x:"))
                break
            except ValueError:
                print("Error! Bookmark not defined.")

        # Receive the coordinate y as input and validate for errors
        while(True):
            try:
                y = float(input("Enter y:"))
                break
            except ValueError:
                print("Error! Bookmark not defined.")

        if checkPointLies(x1, y1, x2, y2, x, y):
            print("True")
        else:
            print("False")
        response = input("Do you want to continue: ")
    print("Goodbye!")

# We define the name of the main function.
if __name__ == "__main__":
    main()
```
